```python
corr_matrix = df.corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, cmap='viridis', linewidth=0.5, vmin=-1, vmax=1, mask=mask)
plt.show()
```

```python
corr_target = corr_matrix['diagnosis']
sns.heatmap(corr_target.to_frame(), cmap='viridis', linewidth=0.5, vmin=-1, vmax=1, annot=True, fmt=".2f")
plt.show()
```

```python
sorted_corr_target = corr_target.abs().sort_values(ascending=False)[1:]
sns.barplot(x=sorted_corr_target.values, y=sorted_corr_target.index, orient='h')
plt.show()
```

```python
top_10 = sorted_corr_target[0:10]
sns.barplot(x=top_10.index, y=top_10.values)
plt.xticks(rotation=90)
plt.title('Top 10 Target Correlated')
plt.grid()
plt.ylim(0.5, 0.8)
plt.show()
```

```python
sorted_for_pairplot = corr_target.abs().sort_values(ascending=False)
sns.pairplot(df[sorted_for_pairplot.index[:5]], hue='diagnosis', palette={0: 'green', 1: 'red'}, markers=['o', 'd'])
plt.show()
```

```python
features = sorted_for_pairplot[1:5].index

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

for i, feature in enumerate(features):
  sns.histplot(df[feature], kde=True, ax=axes[i])
  axes[i].set_title(f"{feature} Distribution")
  axes[i].set_xlabel(feature)
  axes[i].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
def calculate_probability(feature, threshold):
  Malignant = len(df[(df[feature] > threshold) & (df['diagnosis'] == 1)])
  Benign = len(df[(df[feature] > threshold) & (df['diagnosis'] == 0)])
  total = Malignant + Benign
  if total == 0:
    return "No Available Data"
  probability_M = round((Malignant / total) * 100, 2)
  probability_B = round((Benign / total) * 100, 2)
  print(f"{Malignant} Probability of Malignant: {probability_M}%")
  #print(f"{Benign} Probability of Benign: {probability_B}%")
  return probability_M
```

```python
perimeter_insight = calculate_probability('perimeter_mean', 100) - calculate_probability('perimeter_mean', 80)
print(perimeter_insight)
```

```python
concave_point_insight = calculate_probability('concave points_worst', 0.2) - calculate_probability('concave points_worst', 0.1)
print(concave_point_insight)
```

```python
radiust_worst_insight = calculate_probability('radius_worst', 20) - calculate_probability('radius_worst', 10)
print(radiust_worst_insight)
```

```python
concave_point_insight = calculate_probability('concave points_mean', 0.05) - calculate_probability('concave points_mean', 0.01)
print(concave_point_insight)
```

```python
area_insight = calculate_probability('area_mean', 700)
print(area_insight)
```

```python
compactness_mean_insight = calculate_probability('compactness_mean', 0.13)
print(compactness_mean_insight)
```

**Insights**
- Increasing the worst perimeter from 80 to 100 leads to 38% more chance of having a malignant tumer
- Increasing concave points_worst from 0.1 to 0.2 leads to 28.5% more chance of having a malignant tumer
- Increasing radius_worst from 10 to 20 leads to 61.5% more chance of having a malignant tumer
- Increasing concave points_mean from 0.01 to 0.05 leads to 45.7% more chance of having a malignant tumer
- Patients with an area_mean of over 700 are 94% likely to have a malignant tumer
- Patients with a compactness mean more than 0.13 are 80% likekly to have a malignant tumer
